```python
x_axis = np.arange(-50, 50)
def fuc(a, b) :
    f_arr = float(a) * x_axis + float(b)
    print(f_arr)
    return f_arr
plt.title("linear function")

l_fuc = fuc(2,100)
plt.plot(x_axis, l_fuc)
```

## Gradient Descent for linear regression

Hypothesis 
$$h_\theta(x)=\theta_0+\theta_1(x)$$

Objective funtion(cost function)
$$
J(\theta)=\frac{1}{2m} \sum_{i=1}^m (h_\theta(x^{(i)})−y^{(i)})^2
$$

GD 
$$
\theta_0^{(t+1)} : = \theta_0^{(t)} - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})
$$
$$
\theta_1^{(t+1)} : = \theta_1^{(t)} - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) x^{(i)}
$$

```python
x_axis = np.arange(-50, 50)
y = pointset

# y list for graph 
list_th0 = []
list_th1 = []
list_J = [] 

print(x_axis)
print(pointset)

## hypothesis
def h(th0, th1):
    hypo = th0 + th1*(x_axis)
    return hypo

## cost function
def J(th0, th1):
  cost = (1/2*len(x_axis))*np.sum((h(th0, th1)-y)**2)
  return cost
    
##GD algorithm
itr = np.arange(5000)+1
th0 = -100
th1 = 1
a = 0.002

for iteration in np.ndenumerate(itr):
    print("iteration : ", iteration)
    print(th0, th1)
    print("cost:",J(th0, th1))

    temp0 = th0 - a *(1/len(x_axis))*np.sum((h(th0, th1)-y))
    temp1 = th1 - a *(1/len(x_axis))*np.sum((h(th0, th1)-y)*(x_axis))

    #print(h(th0, th1))

    th0 = temp0
    th1 = temp1

    list_th0.append(th0)
    list_th1.append(th1)
    list_J.append(J(th0,th1))
```

```python
#theta0, theta1
plt.plot(itr, np.array(list_th0))
plt.plot(itr, np.array(list_th1))
plt.show()

#J
plt.plot(itr, np.array(list_J) )
plt.show()
```
