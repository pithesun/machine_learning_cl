```python
import numpy as np
import random 
import matplotlib.pyplot as plt
```

```python
#INPUT DATA

## linear function 수식을 받아와서 그리기
x_axis = np.arange(-50, 50)
def fuc(a, b) :
    f_arr = int(a) * x_axis + int(b)
    print(f_arr)
    return f_arr
plt.title("linear function")
l_fuc = fuc(2,100)
plt.plot(x_axis, l_fuc)
```

```python
#linear function에 대해 정규분포된 난수 생성

#가우시안 분포 200개
g_arr = np.random.randn(100)*10
print(g_arr)
pointset = l_fuc + g_arr
print(pointset)
plt.scatter(x_axis, pointset)
```
